# Otto商品分类——数据探索&特征工程

Otto数据集是著名电商Otto提供的一个多类商品分类问题，类别数=9. 每个样本有93维数值型特征（整数，可能表示某种事件发生的次数，已经进行过脱敏处理）。 
竞赛官网：https://www.kaggle.com/c/otto-group-product-classification-challenge/data

第一名：https://www.kaggle.com/c/otto-group-product-classification-challenge/discussion/14335

第二名：http://blog.kaggle.com/2015/06/09/otto-product-classification-winners-interview-2nd-place-alexander-guschin/

```python
# 首先 import 必要的模块
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

## 读取数据

```python
dpath='./data/'
train=pd.read_csv(dpath+"Otto_train.csv")
train.head()
```

```python
train.info()
```

93个整数型特征

```python
## 各属性的统计特性
train.describe()
```

## 标签的分布
查看各类样本的分布是否均衡

```python
sns.countplot(train.target)  #对离散型的特征化直方图
plt.xlabel("target")
plt.ylabel("number of occurrences")
```

各类样本不太均衡，交叉验证对分类任务默认采用StratifiedKFlod,在每折采样时根据样本比例采样。

## 各特征的分布
猜测不同特征的分布差不太多，所以只观察一个特征feat_1的分布

```python
sns.countplot(train.feat_1)
plt.xlabel("feat_1")
plt.ylabel("number of occurrence")
```

**注：特征值大部分是0（稀疏），是长尾分布，可以考虑log(x+1)变换，减弱长尾分布中大特征值的影响**

## 特征之间的相关系数

```python
# get the names of all the columns
cols=train.columns

#Calculates the pearson co-efficient for all combination
#相关系数大于0.5的为强相关
feat_corr=train.corr().abs()
plt.subplots(figsize=(13,9))
sns.heatmap(feat_corr,annot=True)

# Mask unimportant features
sns.heatmap(feat_corr,mask=feat_corr<1,cbar=False)
plt.show()
```

```python
# Set the threshold to select only highly correlated attributes
threshold=0.5
# List of pairs along with correlation above threshold
corr_list=[]

size=feat_corr.shape[0]
# Search for the highly correlated pairs
for i in range(0,size):
    for j in range(i+1,size):
        if feat_corr.iloc[i,j]>=threshold or feat_corr.iloc[i,j]<=-threshold:
            corr_list.append([feat_corr.iloc[i,j],i,j])  #store correlation and columns index
        
corr_list=sorted(corr_list,key=lambda x: abs(x[0]),reverse=True)

for v,i,j in corr_list:
    print("%s and %s = %.2f"%(cols[i],cols[j],v))
            
```

## 特征工程

1、 取对数log1p  
2、tf-idf

### 分开特征和标签

```python
y_train=train['target']
X_train=train.drop(['target','id'],axis=1)

fea_columns=X_train.columns
```

### feat编码：log(x+1)
log(x+1)降低长尾分布中大数值的影响

```python
X_train_log=np.log1p(X_train)

#重新组成DataFrame
fea_names=fea_columns+'_log'
X_train_log=pd.DataFrame(columns=fea_names,data=X_train_log.values)

X_train_log.head()
```

### feat编码：TF-IDF
原始特征feat_x看起来像计数特征，类似文本分析中词频特征的处理，TF-IDF可以突出对特别类别有贡献的低频词。  
这里原始特征已经是计数特征了，直接调用TfidfTransformer，将计数特征变成TF-IDF   
如果输入是原始文本，需要将计数功能（TF）和IDF功能集中在一起，用TfidfVectorizer


```python
#transform counts to TF-IDF features
from sklearn.feature_extraction.text import TfidfTransformer

tfidf=TfidfTransformer()
X_train_tfidf=tfidf.fit_transform(X_train).toarray()

#重新组成DataFrame
fea_names=fea_columns+'_tfidf'
X_train_tfidf=pd.DataFrame(columns=fea_names,data=X_train_tfidf)

X_train_tfidf.head()
```

### 数据预处理
由于数据极度稀疏，数据缩放应采用MinMaxScaler，使得变换后的数据继续保持稀疏。（StandardScale使输入特征变为均值为0，方差为1的正态分布）


```python
#对原始数据放缩
from sklearn.preprocessing import MinMaxScaler

feat_name_org=X_train.columns
#构造输入特征的标准化器
minmax_scaler_org=MinMaxScaler()

X_train=minmax_scaler_org.fit_transform(X_train)
```

```python
#对log数据放缩
feat_name_log=X_train_log.columns

minmax_scaler_log=MinMaxScaler()
X_train_log=minmax_scaler_log.fit_transform(X_train_log)
```

```python
#对tf-idf数据进行放缩
feat_name_tfidf=X_train_tfidf.columns

minmax_scaler_tfidf=MinMaxScaler()
X_train_tfidf=minmax_scaler_tfidf.fit_transform(X_train_tfidf)
```

```python
#保存原始特征
y=pd.Series(data=y_train,name='target')
#print(y)
train_org=pd.concat([pd.DataFrame(columns=feat_name_org,data=X_train),y],axis=1)
#train_org.head()
train_org.to_csv(dpath+"Otto_FE_train_org.csv")
```

```python
#保存log特征变换结果

train_log=pd.concat([pd.DataFrame(columns=feat_name_log,data=X_train_log),y],axis=1)
train_log.to_csv(dpath+"Otto_FE_train_log.csv")
```

```python
#保存tf-idf特征变换结果
train_tfidf=pd.concat([pd.DataFrame(columns=feat_name_tfidf,data=X_train_tfidf),y],axis=1)
train_tfidf.to_csv(dpath+"Otto_FE_train_tfidf.csv")
```

### 保存特征编码过程中用到的模型，用于后续对测试数据的特征进行编码

```python
import pickle as pk
pk.dump(tfidf,open("./model/tfidf.pkl",'wb'))
pk.dump(minmax_scaler_org,open("./model/minmax_scaler_org.pkl",'wb'))
pk.dump(minmax_scaler_log,open("./model/minmax_scaler_log.pkl",'wb'))
pk.dump(minmax_scaler_tfidf,open("./model/minmax_scaler_tfidf.pkl",'wb'))
```

# Logistic回归

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

import matplotlib.pyplot as plt
```

## 读取数据

```python
#读取数据
dpath='./data/'
train=pd.read_csv(dpath+"Otto_FE_train_org.csv")
train.head()
```

## 准备数据

```python
y_train_org=train['target']
X_train_org=train.drop(['target'],axis=1)
X_train_org=X_train_org.iloc[:,1:]
X_train_org.head()
```

## 默认参数的LogistRegression

```python
lr=LogisticRegression()
loss=cross_val_score(lr,X_train_org,y_train_org,cv=3,scoring='neg_log_loss')
print('logloss of each fold is :',-loss)
print('cv logloss is:',np.mean(-loss))
```

## LogisticRegression + GridSearchCV
logistic回归需要调整的超参数有C和正则函数penalty（L2/L1）
目标函数为：$$J=C*sum(logloss(f(X_i),y_i))+penalty$$
在sklearn框架下，不同学习器的参数调整步骤相同：  
1、设置参数搜索范围  
2、生成学习器实例  
3、生成GridSearchCv的实例  
4、调用GridSearchCV的fit方法

```python
penaltys=['l1','l2']
Cs=[0.1,1,10,100,1000]
tuned_parameters=dict(penalty=penaltys,C=Cs)

lr=LogisticRegression(solver='liblinear')
grid=GridSearchCV(lr,tuned_parameters,cv=3,scoring='neg_log_loss',return_train_score=True)
grid.fit(X_train_org,y_train_org)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
# plot CV误差曲线
print(grid.cv_results_.keys())
test_means=grid.cv_results_['mean_test_score']
test_stds=grid.cv_results_['std_test_score']
train_means=grid.cv_results_['mean_train_score']
#print(grid.cv_results_)
train_stds=grid.cv_results_['std_train_score']

#plot resutls
n_Cs=len(Cs)
n_penaltys=len(penaltys)
test_scores=np.array(test_means).reshape(n_Cs,n_penaltys)
train_scores=np.array(train_means).reshape(n_Cs,n_penaltys)
test_stds=np.array(test_stds).reshape(n_Cs,n_penaltys)
train_stds=np.array(train_stds).reshape(n_Cs,n_penaltys)

x_axis=np.log10(Cs)
for i,value in enumerate(penaltys):
    plt.errorbar(x_axis,-test_scores[:,i],yerr=test_stds[:,i],label=penaltys[i]+'Test')
    plt.errorbar(x_axis,-train_scores[:,i],yerr=train_stds[:,i],label=penaltys[i]+'Train')
    
plt.legend()
plt.xlabel('C')
plt.ylabel('logloss')
plt.show()
```

## 保存模型，用于后续测试


```python
import pickle
pickle.dump(grid.best_estimator_,open('./model/Otto_L1_org.pkl','wb'))
```
