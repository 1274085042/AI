# Bikeshare数据集上的数据探索

1、	任务描述  
请在Capital Bikeshare （美国Washington, D.C.的一个共享单车公司）提供的自行车数据上进行回归分析。根据每天的天气信息，预测该天的单车共享量。

原始数据集地址：http://archive.ics.uci.edu/ml/datasets/Bike+Sharing+Dataset  
1)	文件说明  
day.csv: 按天计的单车共享次数（使用该文件）  
hour.csv: 按小时计的单车共享次数（无需理会）  
readme：数据说明文件
  
2)	字段说明  
Instant：记录号  
Dteday：日期  
Season：季节（1=春天、2=夏天、3=秋天、4=冬天）  
yr：年份，(0: 2011, 1:2012)  
mnth：月份( 1 to 12)  
hr：小时 (0 to 23)  （只在hour.csv有，作业忽略此字段）  
holiday：是否是节假日  
weekday：星期中的哪天，取值为0～6  
workingday：是否工作日  
1=工作日 （是否为工作日，1为工作日，0为非周末或节假日）  
weathersit：天气（1：晴天，多云 2：雾天，阴天 3：小雪，小雨）  
temp：气温摄氏度  
atemp：体感温度  
hum：湿度  
windspeed：风速  
casual：非注册用户个数  
registered：注册用户个数  
cnt：给定日期（天）时间（每小时）总租车人数，响应变量y （cnt = casual + registered）

casual、registered和cnt三个特征均为要预测的y，这里只需对cnt进行预测

```python
df=pd.read_csv("day.csv")
df.head()
```

```python
df.describe()
```

## 离散特征的分布

```python
categorical_features=['season','mnth','weathersit','weekday']
for c in categorical_features:
    print("%s 属性不同取值出现的次数："%(c))
    print(df[c].value_counts())
    df[c]=df[c].astype('object')
```

类别型特征的取值不多时

# Bikeshare数据集上的特征工程

## 类别型特征编码
对类别型特征进行独热编码

```python
categorical_features=['season','mnth','weathersit','weekday']
for c in categorical_features:
    df[c]=df[c].astype('object')

train_cat=df[categorical_features]
train_cat=pd.get_dummies(train_cat)
train_cat.head()
```

## 数值型特征标准化
对数值型特征进行标准化/MinMaxScaler，去量纲

```python
from sklearn.preprocessing import MinMaxScaler

minmax=MinMaxScaler()
numberical_features=['temp','atemp','hum','windspeed']
scale=minmax.fit(df[numberical_features])
train_num=scale.transform(df[numberical_features])
#print(train_num.mean(axis=0))
X_train_num=pd.DataFrame(data=train_num,columns=numberical_features,index=df.index)
X_train_num.head()
```

```python
# Join categorical and numerical features
X_train=pd.concat([X_train_num,train_cat,df['holiday'],df['workingday']],axis=1)
X_train.head()
```

```python
FE_train=pd.concat([df['instant'],X_train,df['yr'],df['cnt']],axis=1)
FE_train.to_csv('FE_train.csv')
FE_train.head()
```

```python
FE_train.info()
```

# 模型选择

```python
#准备数据
from sklearn.model_selection import train_test_split

data=pd.read_csv('FE_train.csv')
y=data['cnt']
X=data.drop('cnt',axis=1)

features=X.columns

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=33,test_size=0.2)

print(X_train.shape)
print(X_test.shape)
```

## 线性回归模型

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(X_train,y_train)

lr_w=pd.DataFrame({"columns":features,"coef":lr.coef_})
lr_w.sort_values(by='coef')
```

### 评估

```python
from sklearn.metrics import r2_score
y_pre=lr.predict(X_test)
print("The r2 score of LinearRegression on test is ",r2_score(y_test,y_pre))
```

## 岭回归

```python
from sklearn.linear_model import RidgeCV

ridge=RidgeCV()
ridge.fit(X_train,y_train)
lr_w.shape
ridge_w=pd.DataFrame({'columns':features,'lr':lr.coef_,'ridge':ridge.coef_})
ridge_w.sort_values(by='ridge')
```

### 评估

```python
y_pre=ridge.predict(X_test)
print("The r2 score of Ridge on test is ",r2_score(y_test,y_pre))
```

## Lasso

```python
from sklearn.linear_model import LassoCV

alphas=np.logspace(-5,2,10)
print(alphas)
lasso=LassoCV(alphas=alphas,max_iter=10000)
lasso.fit(X_train,y_train)

lasso_w=pd.DataFrame({"columns":features,"lr":lr.coef_,"ridge":ridge.coef_,"lasso":lasso.coef_})
lasso_w.sort_values(by='lasso')
```

```python
print(lasso.alpha_)
```

### 评估

```python
y_pre=lasso.predict(X_test)
print ('The r2 score of LassoCV on test is', r2_score(y_test, y_pre))
```
